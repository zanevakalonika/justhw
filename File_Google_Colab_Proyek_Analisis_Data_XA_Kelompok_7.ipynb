```python
# Anggota Kelompok : Chelsea/09, Michelle/21, Rachel/27, Neva/34
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

data = pd.read_excel('Sumber_Sampah_SIPSN.xlsx', header= 1)
data
```

```python
total_smbr_perprov = data.groupby(['Tahun', 'Provinsi'])['Rumah Tangga(ton)', 'Perkantoran(ton)', 'Pasar(ton)', 'Perniagaan(ton)', 'Fasilitas Publik(ton)', 'Kawasan(ton)', 'Lain(ton)'].sum()
total_smbr_perprov
```

```python
avg_total = data.groupby('Provinsi')['Total(ton)'].mean()
avg_total
```

```python
prov_smbr_max = data.loc[data.groupby('Tahun')['Total(ton)'].idxmax()][['Tahun', 'Provinsi', 'Total(ton)']]
prov_smbr_max
```

```python
prov_smbr_min = data.loc[data.groupby('Tahun')['Total(ton)'].idxmin()][['Tahun', 'Provinsi', 'Total(ton)']]
prov_smbr_min
```

```python
to = data.groupby(['Tahun', 'Provinsi'])['Total(ton)'].sum().unstack()

to.plot(kind='bar', stacked=False, figsize=(16, 9))
plt.title('Grafik Total Sumber Sampah per - Provinsi & Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Total Sumber Sampah (ton)')
plt.legend(title='Provinsi', loc='lower left', bbox_to_anchor= (1, 0))

plt.show()
```

```python
to = data.groupby(['Tahun', 'Provinsi'])['Rumah Tangga(ton)', 'Perkantoran(ton)', 'Pasar(ton)', 'Perniagaan(ton)', 'Fasilitas Publik(ton)', 'Kawasan(ton)', 'Lain(ton)'].sum().unstack()

to.plot(kind='bar', stacked=False, figsize=(16, 9))
plt.title('Grafik Total Sumber Sampah per - Provinsi & Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Total Sumber Sampah (ton)')
plt.legend(title='Provinsi', loc='lower left', bbox_to_anchor= (1, 0))

plt.show()
```

```python
avg_all = data.groupby('Provinsi')['Total(ton)'].mean()

colors = []
for category in avg_all:
    if category <= 2000:
        colors.append('green')
    elif category <= 4000:
        colors.append('orange')
    else:
        colors.append('red')

KangKungChef = avg_all.plot(kind="bar", figsize=(16,9), color=colors)

legend_labels = {
    'green': 'HIJAU',
    'orange': 'ORANYE',
    'red': 'MERAH'
}
handles = [plt.Rectangle((0,0),1,1, color=color) for color in legend_labels.keys()]
labels = [label for label in legend_labels.values()]
KangKungChef.legend(handles, labels, title='Kategori', loc='upper right', bbox_to_anchor=(1, 1))

plt.title('Kategori Rata - Rata Total Sumber Sampah per - Provinsi & Tahun')
plt.xlabel('Provinsi')
plt.ylabel('Rata - Rata Sumber Sampah (ton)')

plt.show()
```
